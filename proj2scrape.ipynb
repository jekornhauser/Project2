```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
bank_name = []
country = []
assets_in_bil_us = []
bal_sheet_date = []
```

```python
urls= ['https://web.archive.org/web/20120626023412/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20121225122536/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20131111012357/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20140209050456/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20140701123856/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20150512003339/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20151104133716/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20160504071841/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20161120235852/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20170515081439/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20180205184632/https://www.relbanks.com/worlds-top-banks/assets',\
'https://web.archive.org/web/20181224170514/https://www.relbanks.com/worlds-top-banks/assets/']
```

```python
for i in urls:
    browser.visit(i)

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    table = soup.find('table', class_='font8c')

    rows = table.find_all('tr')

    for row in rows:
        info = row.find_all('td')
        bank_name.append(info[1].text)
        country.append(info[2].text)
        assets_in_bil_us.append(info[3].text)
        bal_sheet_date.append(info[4].text)
```

```python
print(bank_name)
print(country)
print(assets_in_bil_us)
print(bal_sheet_date)
```

```python
import pandas as pd
```

```python
data = {'Bank': bank_name, 'Country': country, 'Total Assets': assets_in_bil_us, 'Balance Sheet Date': bal_sheet_date}
banksdf = pd.DataFrame(data, columns = ['Bank','Country','Total Assets','Balance Sheet Date'])
banksdf.head()
```

```python
#replace .s with /s for consistency
banksdf['Balance Sheet Date'] = banksdf['Balance Sheet Date'].str.replace('.','/',regex=False)
```
